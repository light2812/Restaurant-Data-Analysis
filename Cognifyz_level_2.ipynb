# **Task 1**

Analyze the distribution of aggregate ratings and determine the most common rating range.

```python
# Remove rows with Aggregate rating of 0 or "Not Rated"
cleaned_data = data[(data['Aggregate rating'] != 0) & (data['Rating text'] != 'Not rated')]

plt.figure(figsize=(10, 6))
sns.histplot(cleaned_data['Aggregate rating'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.show()

rating_ranges = {
    '1-2': (1, 2),
    '2-3': (2, 3),
    '3-4': (3, 4),
    '4-5': (4, 5)
}

rating_counts = {range_label: cleaned_data[(cleaned_data['Aggregate rating'] >= low) & (cleaned_data['Aggregate rating'] < high)].shape[0]
                 for range_label, (low, high) in rating_ranges.items()}

# Most common rating range
most_common_range = max(rating_counts, key=rating_counts.get)
print(f"The most common rating range is: {most_common_range} with {rating_counts[most_common_range]} restaurants.")
```

# **Task 3**

Plot the locations of restaurants on a map using longitude and latitude coordinates.

```python
data_cleaned = data[(data['Latitude'] != 0) & (data['Longitude'] != 0)]
data_cleaned = data_cleaned[(data_cleaned['Latitude'] >= -90) & (data_cleaned['Latitude'] <= 90)]
data_cleaned = data_cleaned[(data_cleaned['Longitude'] >= -180) & (data_cleaned['Longitude'] <= 180)]
```

```python
import folium
from folium.plugins import MarkerCluster

# Create a map centered around the median latitude and longitude
m = folium.Map(location=[data_cleaned['Latitude'].median(), data_cleaned['Longitude'].median()], zoom_start=5)

# Initialize MarkerCluster
marker_cluster = MarkerCluster().add_to(m)

# Add restaurant markers to the cluster
for idx, row in data_cleaned.iterrows():
    folium.Marker([row['Latitude'], row['Longitude']], popup=row['Restaurant Name']).add_to(marker_cluster)


```

Identify any patterns or clusters of
restaurants in specific areas.

```python
from sklearn.cluster import KMeans
data_clean = data.dropna(subset=['Latitude', 'Longitude'])

# Extract latitude and longitude
coords = data_clean[['Latitude', 'Longitude']]

# Normalize the coordinates (scaling to a common range)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
coords_scaled = scaler.fit_transform(coords)

kmeans = KMeans(n_clusters=5, random_state=0)
data_clean['Cluster'] = kmeans.fit_predict(coords_scaled)


plt.figure(figsize=(10, 6))
sns.scatterplot(x=data_clean['Longitude'], y=data_clean['Latitude'], hue=data_clean['Cluster'], palette="viridis", s=100, alpha=0.7)
plt.title("Clusters of Restaurants Based on Location", fontsize=16)
plt.xlabel("Longitude", fontsize=12)
plt.ylabel("Latitude", fontsize=12)
plt.legend(title="Cluster")
plt.show()
```

Analyze the ratings and popularity of different restaurant chains.

```python
# Step 1: Identify restaurant chains (names that appear more than once)
restaurant_counts = cleaned_data['Restaurant Name'].value_counts()
restaurant_chains = restaurant_counts[restaurant_counts > 1]

# Step 2: Filter the dataset to include only restaurant chains
chains_data = cleaned_data[cleaned_data['Restaurant Name'].isin(restaurant_chains.index)]

# Step 3: Calculate the average rating and total votes for each restaurant chain
chain_summary = chains_data.groupby('Restaurant Name').agg(
    average_rating=('Aggregate rating', 'mean'),
    total_votes=('Votes', 'sum')
)

# Step 4: Sort chains by total votes (popularity) and display the top 10
top_chains_by_votes = chain_summary.sort_values(by='total_votes', ascending=False).head(10)
top_chains_by_rating = chain_summary.sort_values(by='average_rating', ascending=False).head(10)

print("Top 10 Restaurant Chains by Popularity (Votes):")
print(top_chains_by_votes)

print("\nTop 10 Restaurant Chains by Average Rating:")
print(top_chains_by_rating)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_chains_by_votes.index, y=top_chains_by_votes['total_votes'],hue=top_chains_by_votes.index ,palette='viridis')
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 Restaurant Chains by Popularity (Votes)')
plt.xlabel('Restaurant Chain')
plt.ylabel('Total Votes')
plt.tight_layout()
plt.show()


```
