# **TASK 1**


Determine the top three most
common cuisines in the dataset.


```python
top3_cuisines = data['Cuisines'].value_counts().head(3)
print("Top 3 Cuisines:\n", top3_cuisines)
```

```python
# Split multi-cuisine values and create a new list because from above , we can see that many entries contain multiple cuisines available in a restaurant.
cuisines_split = data['Cuisines'].dropna().str.split(',').explode().str.strip()

# Get the top 3 cuisines
top_cuisines = cuisines_split.value_counts().head(3)
print(top_cuisines)
```

We can see that these are the top 3 Cuisines.

................................................................................

Calculate the percentage of
restaurants that serve each of the top
cuisines.

```python
# Split the cuisines column by commas and get the count of each cuisine
cuisines_list = data['Cuisines'].str.split(',').explode().value_counts()

# Calculate the total number of restaurants
total_restaurants = len(data)

# Calculate the percentage of restaurants serving each cuisine
cuisines_percentage = (cuisines_list / total_restaurants) * 100

top_cuisines = cuisines_percentage.head(3)
print("Percentages of top 3 cuisines:")
print(top_cuisines)
```

```python
plt.figure(figsize=(10, 6))
top_cuisines.plot(kind='bar', color='skyblue')

# Add labels and title
plt.title('Top Cuisines by Percentage of Restaurants')
plt.xlabel('Cuisine')
plt.ylabel('Percentage of Restaurants')
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

# **TASK 2**

Identify the city with the highest number of restaurants in the dataset.


```python
city_maxrest = data['City'].value_counts().idxmax()
top_city_count = data['City'].value_counts().max()
print(f"The city with the highest number of restaurants is: {city_maxrest} with {top_city_count} Restraunts")
```

................................................................................

Calculate the average rating for restaurants in each city.


```python
# Remove rows where Aggregate rating for restaurant is 0 where corresponding Rating text is 'Not rated'
data_filtered = data[(data['Aggregate rating'] > 0) & (data['Rating text'] != 'Not rated')]

# Calculate the average rating for each city
average_rating_per_city = data_filtered.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False)

# Display the result
print(average_rating_per_city.head(10))
```

................................................................................

Determine the city with the highest average rating.

```python
# Determine the city with the highest average rating
city_highest_avg_rating = average_rating_per_city.idxmax()
highest_avg_rating = average_rating_per_city.max()

print(f"The city with the highest average rating is: {city_highest_avg_rating} with an average rating of {highest_avg_rating:.2f}")
```

# **TASK 3**

Create a histogram or bar chart to visualize the distribution of price ranges among the restaurants.



```python
price_range_counts = data['Price range'].value_counts().sort_index()
plt.figure(figsize=(8, 5))
sns.barplot(x=price_range_counts.index, y=price_range_counts.values, palette="viridis",hue=price_range_counts.index, dodge=False)

for i, value in enumerate(price_range_counts.values):
    plt.text(x=i, y=value + 10, s=str(value), ha='center', fontsize=10, color='black')

plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")
plt.title("Distribution of Price Ranges Among Restaurants")
plt.xticks(ticks=range(len(price_range_counts.index)), labels=price_range_counts.index)
plt.show()
```

................................................................................

Calculate the percentage of restaurants in each price range category.

```python
price_range_percentages = (price_range_counts / total_restaurants) * 100

print("Percentage of Restaurants in Each Price Range:")
for price_range, percentage in price_range_percentages.items():
    print(f"Price Range {price_range}: {percentage:.2f}%")
```

# **TASK 4**

Determine the percentage of restaurants
that offer online delivery.

```python
online_delivery_counts = data['Has Online delivery'].value_counts()
percentage_online_delivery = (online_delivery_counts['Yes'] / total_restaurants) * 100
print(f"Percentage of restaurants offering online delivery: {percentage_online_delivery:.2f}%")
```

A quarter of restaurants offer online delivery, suggesting it may not be a standard feature for many restaurants.

................................................................................

Compare the average ratings of restaurants
with and without online delivery.

```python
# Filter out rows where rating is 0.0 or 'Not rated' to avoid skewness
filtered_data = data[(data['Aggregate rating'] > 0) & (data['Rating text'] != 'Not rated')]

online_delivery_avg_rating = filtered_data[filtered_data['Has Online delivery'] == 'Yes']['Aggregate rating'].mean()
no_online_delivery_avg_rating = filtered_data[filtered_data['Has Online delivery'] == 'No']['Aggregate rating'].mean()

print(f"Average rating for restaurants offering online delivery: {online_delivery_avg_rating:.2f}")
print(f"Average rating for restaurants not offering online delivery: {no_online_delivery_avg_rating:.2f}")
```

Restaurants without online delivery tend to have slightly higher ratings on average. This may be due to price ranges of Restaurants.
