# **Task 1**

Analyze the text reviews to identify the most common positive and negative keywords

```python
filtered_data = data[data['Rating text'].isin(['Excellent', 'Very Good', 'Good', 'Poor','Average'])].copy()
```

```python
rating_counts = filtered_data['Rating text'].value_counts()

# Display the result
print("Frequency of Rating Text:")
print(rating_counts)
```

```python
sentiment_map = {
    'Excellent': 'Positive',
    'Very Good': 'Positive',
    'Good': 'Positive',
    'Average': 'Negative',#can also be neutral, but we choose it to be negative
    'Poor': 'Negative',

}

filtered_data['Sentiment'] = filtered_data['Rating text'].map(sentiment_map)
sentiment_counts = filtered_data['Sentiment'].value_counts()
plt.figure(figsize=(8, 5))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values)

# Annotate bars with count values
for i, value in enumerate(sentiment_counts.values):
    plt.text(x=i, y=value + 10, s=str(value), ha='center', fontsize=12)

plt.xlabel("Sentiment")
plt.ylabel("Number of Reviews")
plt.title("Distribution of Sentiments Based on Rating Text")
plt.show()

# Display the counts of each sentiment category
print("Sentiment Distribution:")
print(sentiment_counts)
```

```python
from wordcloud import WordCloud

# Step 1: Filter the data for positive and negative sentiments
positive_reviews = filtered_data[filtered_data['Sentiment'] == 'Positive']['Rating text']
negative_reviews = filtered_data[filtered_data['Sentiment'] == 'Negative']['Rating text']

# Step 2: Join the reviews into a single string for each sentiment
positive_text = ' '.join(positive_reviews)
negative_text = ' '.join(negative_reviews)

# Step 3: Generate the word clouds
positive_wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='Blues').generate(positive_text)
negative_wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='Reds').generate(negative_text)

# Step 4: Plot the word clouds
plt.figure(figsize=(12, 8))

# Positive sentiment word cloud
plt.subplot(1, 2, 1)
plt.imshow(positive_wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Positive Sentiments')

# Negative sentiment word cloud
plt.subplot(1, 2, 2)
plt.imshow(negative_wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Negative Sentiments')

plt.show()
```

Calculate the average length of reviews and explore if there is a relationship between review length and rating.

```python
#We dont use cleaned data because 'not rated' also has word length

data['Review Length'] = data['Rating text'].apply(lambda x: len(str(x)))
average_review_length = data['Review Length'].mean()
print(f"Average review length: {average_review_length:.2f} characters")


plt.figure(figsize=(8, 6))
sns.scatterplot(x='Review Length', y='Aggregate rating', data=data, alpha=0.6)
plt.xlabel('Review Length (in characters)')
plt.ylabel('Aggregate Rating')
plt.title('Review Length vs. Rating')
plt.show()

correlation = data[['Review Length', 'Aggregate rating']].corr()
print("\nCorrelation between review length and rating:")
print(correlation)
```

Rating of 4-5--->Review length of 9(EXCELLENT)
Rating of 3.5-4-->Review length of 4(GOOD)
Rating of 2.5-3.5--->Review length of 7(AVERAGE)

# **Task 3**

Analyze if there is a relationship between the price range and the availability of online delivery and table booking.

```python
# Calculate counts for Online Delivery and Table Booking for each Price Range
online_delivery_counts = data.groupby(['Price range', 'Has Online delivery']).size().unstack(fill_value=0)
table_booking_counts = data.groupby(['Price range', 'Has Table booking']).size().unstack(fill_value=0)

# Displaying the counts of Online Delivery and Table Booking for each Price Range
print("Online Delivery Counts by Price Range:")
print(online_delivery_counts)

print("\nTable Booking Counts by Price Range:")
print(table_booking_counts)
```

```python
price_delivery_table = data.groupby(['Price range', 'Has Online delivery'])['Has Online delivery'].count().unstack(fill_value=0)
price_table_booking = data.groupby(['Price range', 'Has Table booking'])['Has Table booking'].count().unstack(fill_value=0)

plt.figure(figsize=(10, 6))
price_delivery_table.plot(kind='bar', stacked=True, figsize=(10, 6), color=['skyblue', 'orange'])
plt.title("Online Delivery Availability by Price Range", fontsize=16)
plt.xlabel("Price Range", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=0)
plt.legend(title="Online Delivery")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
price_table_booking.plot(kind='bar', stacked=True, figsize=(10, 6), color=['lightgreen', 'purple'])
plt.title("Table Booking Availability by Price Range", fontsize=16)
plt.xlabel("Price Range", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=0)
plt.legend(title="Table Booking")
plt.show()
```

Online Delivery is more prevalent in mid-range restaurants (Price Range 2 and 3), while high-end (Range 4) and budget (Range 1) restaurants are less likely to offer it.


Table Booking is uncommon in budget restaurants but increases with price range. It becomes almost evenly split between availability and non-availability in high-end restaurants (Range 4).

Determine if higher-priced restaurants are more likely to offer these services

```python
# Calculate the percentage of restaurants offering online delivery and table booking in each price range
online_delivery_percent = data.groupby('Price range')['Has Online delivery'].value_counts(normalize=True).unstack() * 100
table_booking_percent = data.groupby('Price range')['Has Table booking'].value_counts(normalize=True).unstack() * 100

# Display the results
print("Percentage of Restaurants Offering Online Delivery by Price Range:")
print(online_delivery_percent)

print("\nPercentage of Restaurants Offering Table Booking by Price Range:")
print(table_booking_percent)
```

```python
online_delivery_percent.plot(kind='bar', stacked=True, figsize=(10, 6), color=['skyblue', 'orange'])
plt.title("Percentage of Restaurants Offering Online Delivery by Price Range", fontsize=16)
plt.xlabel("Price Range", fontsize=12)
plt.ylabel("Percentage", fontsize=12)
plt.xticks(rotation=0)
plt.legend(title="Online Delivery")
plt.show()
```

```python
table_booking_percent.plot(kind='bar', stacked=True, figsize=(10, 6), color=['lightgreen', 'purple'])
plt.title("Percentage of Restaurants Offering Table Booking by Price Range", fontsize=16)
plt.xlabel("Price Range", fontsize=12)
plt.ylabel("Percentage", fontsize=12)
plt.xticks(rotation=0)
plt.legend(title="Table Booking")
plt.show()
```

Higher-priced restaurants are much more likely to offer table booking services, as indicated by the significant increase in percentages for Price Ranges 3 and 4.


Online delivery appears more prevalent in mid-priced restaurants (Price Range 2) than in the highest-priced category, suggesting a different focus on service offerings.
